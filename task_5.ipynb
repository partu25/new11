# Implement Decision Tree(ID3) in python
Uses Information Gain to choose the best feature to split. 

Recursively builds the tree until stopping conditions are met.

1) Calculate Entropy for the dataset.<BR>
2) Calculate Information Gain for each feature. <BR>
3) Choose the feature with maximum Information Gain. <BR>
4) Split dataset into subsets for that feature. <BR>
5) Repeat recursively until: <BR>

All samples in a node have the same label.<BR>
No features are left.<BR>
No data is left.

## Implement ID3 Algo

```python

def entropy(target_col):
    elements, counts = np.unique(target_col, return_counts=True)
    probabilities = counts / counts.sum()
    return -np.sum(probabilities * np.log2(probabilities + 1e-9))  # small value to avoid log(0)

def information_gain(data, feature, target_name):
    total_entropy = entropy(data[target_name])
    vals, counts = np.unique(data[feature], return_counts=True)
    weighted_entropy = np.sum([
        (counts[i] / np.sum(counts)) * entropy(data[data[feature] == vals[i]][target_name])
        for i in range(len(vals))
    ])
    return total_entropy - weighted_entropy

def id3(data, features, target_name):
    # If all target values are the same → return that class
    if len(np.unique(data[target_name])) == 1:
        return np.unique(data[target_name])[0]
    
    # If no features left → return majority class
    if len(features) == 0:
        return data[target_name].mode()[0]
    
    # Select feature with highest information gain
    ig_values = [information_gain(data, feature, target_name) for feature in features]
    best_feature_index = np.argmax(ig_values)
    best_feature = features[best_feature_index]
    
    tree = {best_feature: {}}
    
    # Recursively create branches
    for value in np.unique(data[best_feature]):
        subset = data[data[best_feature] == value].drop(columns=[best_feature])
        subtree = id3(subset, [f for f in features if f != best_feature], target_name)
        tree[best_feature][value] = subtree
    
    return tree
```

```python

# Load heart dataset
data = pd.read_csv("heart.csv")

# Strip whitespace from column names (if any)
data.columns = data.columns.str.strip()

# Features list (excluding target)
features = [col for col in data.columns if col != 'target']

# Build decision tree
tree = id3(data, features, 'target')

# Display the tree
import pprint
pprint.pprint(tree)
```

## Extra: Create Predict Function

```python
def predict(tree, sample):
    if not isinstance(tree, dict):
        return tree
    feature = list(tree.keys())[0]
    branches = tree[feature]
    value = sample.get(feature)
    if value in branches:
        result = branches[value]
        if isinstance(result, dict):
            return predict(result, sample)
        else:
            return result
    else:
        return None
```
